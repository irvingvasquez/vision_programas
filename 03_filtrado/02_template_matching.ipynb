# Template matching

En este notebook veremos una de las características que tiene la convolución

```python
# paquetes necesarios
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import cv2
%matplotlib inline
```

```python
imagen_color = mpimg.imread('../02_geometria/ESCOM2_small.jpg')
#print(imagen_color)
imagen = cv2.cvtColor(imagen_color, cv2.COLOR_BGR2GRAY)
plt.imshow(imagen, cmap='gray')
```

```python
kernel_color = mpimg.imread('cidetec.jpg')
kernel = cv2.cvtColor(kernel_color, cv2.COLOR_BGR2GRAY)
plt.imshow(kernel, cmap='gray')
```

```python
# Operación de convolución 2D entre el filtro y la imagen
img_filtrada = cv2.filter2D(imagen.astype('float32'), -1, kernel.astype('float32'))


minv = np.amin(img_filtrada)
print(minv)

img_filtrada = (img_filtrada - minv)

maxv = np.amax(img_filtrada)
print(maxv)

img_filtrada = (img_filtrada) * (1/maxv)

print(img_filtrada)

#imprimir la magen convolucionada
plt.imshow(img_filtrada)
```
